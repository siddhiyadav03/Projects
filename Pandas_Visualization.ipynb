```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings(action="ignore")
```

```python
df=pd.read_csv('choclate protfolio project - 11.csv')
```

```python
#displays all the data
df
```

```python
#displays first 5 rows
df.head()
```

```python
#returns the number of missing values in the dataset
df.isnull().sum()
```

```python
#returns the label of each column in the datadrame
df.columns
```

```python
#prints information of the dataframe
df.info()
```

```python
#replaces the specified value with a new value
#df.replace()
df["Amount"]=df["Amount"].str.replace("$"," ")
df["Amount"]=df["Amount"].str.replace(",","")
df["Amount"]=df["Amount"].astype("int64")
df
```

```python
df["Cost"]=df["Cost"].str.replace("$"," ")
df["Cost"]=df["Cost"].str.replace(",","")
df["Cost"]=df["Cost"].astype(int)
```

```python
df
```

```python
#replace geography with countries
df.rename(columns={'Geography': 'Countries'},inplace=True)
df
```

```python
#drop a specified column name
#The first argument, ["Profit", "profit %"], specifies the names of the columns to be dropped.
#The second argument, axis=1, specifies that the columns should be dropped along the "columns" axis (i.e., horizontally).
df.drop(['Profit','profit %'],axis=True)
df
```

```python
# to generate descriptive statistics of a Pandas DataFrame object df.
df.describe()
```

```python
#sort the Pandas DataFrame object df in ascending order based on the values in the "Profit" column. 
df.sort_values(by=['Profit'],ascending=True)
```

```python
#unique values of the "Countries" column from the Pandas DataFrame object df. 
df['Countries'].unique()
```

```python
#create a new Pandas DataFrame object df_new by grouping the existing DataFrame object df by the "Countries" column 
#And calculate the sum of the "Profit" column for each group.
df_new=df[['Countries','Profit']].groupby('Countries')['Profit'].sum()
df_new   #the variable df_new,  has the "Countries" column as its index and a single "Profit" column as the summed values for each country.
```

```python
#replace()
df['Profit']=df['Profit'].str.replace('$','')
df['Profit']=df['Profit'].str.replace('(','-')
df['Profit']=df['Profit'].str.replace(')','')
df['Profit']=df['Profit'].str.replace(',','')
df
```

```python
#convert the data type of the "Profit" column from a non-integer type to an integer type using the astype() method.
df['Profit'] = df['Profit'].astype(int)
```

```python
#display the summary information of the Pandas DataFrame object df 
df.info()
```

```python
#sort the Pandas DataFrame object df in ascending order based on the values in the "Profit" column.
df.sort_values(by=['Profit'],ascending=True)
```

```python
#creates a new Pandas DataFrame object data_pivot using the pivot_table() method to summarize the data in the original DataFrame df.
data_pivot=pd.pivot_table(df,index="Countries",values="Amount",aggfunc="sum")
data_pivot
```

```python
#create a new column named "Profit" that represents the profit earned for each transaction.
df["Profit"]=df["Amount"]-df["Cost"]
df
```

```python
#create a new column named "Profit_percent" that represents the percentage of profit earned for each transaction.
df["Profit_percent"]=((df["Amount"]-df["Cost"])/df["Amount"])*100
df
```

```python
#group the rows in df by two columns, "Countries" and "Product", and then calculate the sum of the "Profit" column for each group.
df2=df[['Countries','Profit','Product']].groupby(['Countries','Product'])['Profit'].sum()
df2
```

```python
#group the rows in df by the "Product" column, and then calculate the sum of the "Profit" column for each group.
df3=df[['Product','Profit']].groupby('Product')['Profit'].sum()
df3
```

```python
# select only the rows where the "Profit" column is less than zero.
df_losses=df[df['Profit']<0]
df_losses
```

```python
#sort the DataFrame in ascending order of the "Profit" column, so the products with the largest negative profit will appear at the top of the DataFrame.
df4=df_losses[['Product','Profit']].groupby('Product')['Profit'].sum().sort_values()
df4
```

```python
#summary of the mean profit percentage for each combination of product and country.
data_pt_profit=pd.pivot_table(df,index=("Product"),columns=("Countries"),values=["Profit_percent"],aggfunc=(np.mean))
data_pt_profit.sort_values(by=("Product"),ascending=True)
```

```python
#highlight any cells with negative profit percentages in red.
styled_table = data_pt_profit.style.format('{:.2f}%').applymap(lambda x: 'background-color: red' if x < 0 else '')
styled_table
```

```python
#summary of the total cost for each country.
top_country=pd.pivot_table(df,index="Countries",values="Cost",aggfunc=np.sum)
top_country
```

```python
#summary of the total number of units sold by each sales person.
top_Sales_person=pd.pivot_table(df,index="Sales Person",values="Units",aggfunc=np.sum)
top_Sales_person.head(5)
```

```python
#create a new DataFrame data_numeric that contains only the columns from the original df DataFrame that have a numeric data type.
data_numeric=df.select_dtypes(include="int")
data_numeric
```
